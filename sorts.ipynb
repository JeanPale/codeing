```python
import numpy
import math
import matplotlib.pyplot
import time
```

```python
def Bubble_Sort(your_list):
    my_list=list(your_list)
    done = True
    for j in range(len(my_list)-1):
        for i in range(len(my_list)-1):
            if my_list[i] > my_list[i+1]:
                my_list[i],my_list[i+1] = my_list[i+1],my_list[i]
                done=False
        if done: break
    return my_list
```

```python
x = list(range(10,1000,10))
N = 100
avg_time_values = [] 
max_time_values = []
total_process_time = 0
process_start  = time.time()
for n in x:
    total_time = 0
    max_time = 0
    for k in range(N):
        my_array = numpy.random.rand(n)
        start  = time.time()
        Bubble_Sort(my_array)
        finish = time.time()
        total_time += finish - start
        max_time = max(finish - start, max_time)
    avg_time_values.append(total_time/N)   
    max_time_values.append(max_time)

process_finish  = time.time()
total_process_time += process_finish - process_start

matplotlib.pyplot.plot(x,avg_time_values)
matplotlib.pyplot.plot(x,max_time_values)

    # Расчёт времени всего процесса 

tp_hrs = total_process_time//3600
tp_mnts = (total_process_time - tp_hrs*3600)//60
tp_scnds = total_process_time - tp_hrs*3600 - tp_mnts*60
print("Расчеты заняли всего лишь", int(tp_hrs), "часов", int(tp_mnts), "минут", int(tp_scnds), "секунд!" ) 
    
```

```python
def Radix_Sort(your_list):
    my_list = list(your_list)
    N = int(math.log10(max(my_list)))
    for n in range(N + 1):
        temp_rank_table = [list() for i in range(10)]
        for m in my_list:
            temp_rank_table[(m//10**n)%10].append(m)
        my_list = list()
        for i in temp_rank_table:
            my_list += i
    return my_list      
```

```python
def Merge_Comp(arr_p1, arr_p2):
    new_arr = []
    n1 = len(arr_p1)
    n2 = len(arr_p2)
    i = 0; k = 0
    for m in range(n1 + n2):
        if arr_p1[i] < arr_p2[k]:
            new_arr.append(arr_p1[i])
            i += 1
            if i > n1-1:
                new_arr += arr_p2[k:]
                break
        else:
            new_arr.append(arr_p2[k])
            k += 1
            if k > n2-1:
                new_arr += arr_p1[i:]
                break
    return new_arr
```

```python
def Merge_Sort(your_list):
    my_list = list(your_list)
    n = len(my_list)
    if n == 1:
        return my_list
    arr_p1 = my_list[:n//2]
    arr_p2 = my_list[n//2:]
    return Merge_Comp(Merge_Sort(arr_p1), Merge_Sort(arr_p2))
```
